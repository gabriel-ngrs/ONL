```julia
using ForwardDiff
function aval(f, x, k)
    if k == 0
        return f(x)
    elseif k == 1
        return ForwardDiff.gradient(f, x)
    elseif k == 2
        return ForwardDiff.hessian(f, x)
    else
        print("Erro na vari´avel ordem")
    end
end
```

```julia
using ForwardDiff, LinearAlgebra

function quadratica(A, x, k)
    f(x)=0.5* x'*A*x
    if (k == 0)
        print(" f(x)=")
        return f(x)
    elseif (k == 1)
        print("Gradiente f(x)=")
        return ForwardDiff.gradient(f, x)
    elseif (k == 2)
        print("Hessiana f(x)=")
        return ForwardDiff.hessian(f, x)
    else
        print("Erro na variável ordem")
    end
end
```
